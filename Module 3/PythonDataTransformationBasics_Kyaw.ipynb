```python
import pandas as pd
```

```python
#imporing the dataset "crimestat (1).csv"
nRowsRead = 1000
crime_df = pd.read_csv('crimestat (1).csv', delimiter=',', nrows = nRowsRead, index_col='INC NUMBER')
crime_df.dataframeName = 'crimestat (1).csv'

crime_df.head(10)
```

**Grouping crimes by premise type**

```python
#sorting and grouping by "premise type"
sorted_type = crime_df.sort_values(by=['PREMISE TYPE'])
grouped_type = sorted_type.groupby('PREMISE TYPE')
grouped_type.head(10)
```

```python
#sorting and grouping by zip code
sorted_zipcode = crime_df.sort_values(by=['ZIP'])
grouped_zip = sorted_type.groupby('ZIP')
grouped_zip.head(10)
```

**Looking at types of crimes and where they are most common**

```python
#finding the most common crimes
print(crime_df['UCR CRIME CATEGORY'].value_counts())

#sorting the data frame by crime type
sorted_crime_type = crime_df.sort_values(by=['UCR CRIME CATEGORY'])


```

```python
#grouping the data frame by sorted crime type
grouped_crime_type = sorted_crime_type.groupby('UCR CRIME CATEGORY')
grouped_crime_type.head(10)
```

```python
#finding all crime counts in each zipcode in 2015
grouped_crime_location = crime_df.groupby(['ZIP', 'UCR CRIME CATEGORY'])
crime_counts = grouped_crime_location.size().reset_index(name='Count')

crime_counts
```

```python
#finding the occurences (counts) of three most common crimes in different zipcodes.
three_most_common_crime_df = crime_df[crime_df['UCR CRIME CATEGORY'].isin(['LARCENY-THEFT', 'BURGLARY', 'MOTOR VEHICLE THEFT'])]
```

```python
#grouping the three most common crimes by location
grouped_three_most_crime_location = three_most_common_crime_df.groupby(['UCR CRIME CATEGORY', 'ZIP'])
three_most_crime_counts = grouped_three_most_crime_location.size().reset_index(name='Count')

#sorting the dataframe by crime type and count
sorted_three_most_crime_counts = three_most_crime_counts.sort_values(by=['UCR CRIME CATEGORY', 'Count'], ascending=[True, False])

sorted_three_most_crime_counts

#print(sorted_three_most_crime_counts)
```

**Grouping Violent and Non-Violent Crimes**

```python
#finding the unique crime types(catagories)
list_crime_types = crime_df['UCR CRIME CATEGORY'].unique()
print(list_crime_types)
```

```python
#grouping the non-violent crimes


# filtering non-violent crimes (Theft or Drug Offense) using apply method and lambda function
non_violent_crime_df = crime_df[crime_df['UCR CRIME CATEGORY'].apply(lambda x: 'THEFT' in x.upper() or 'DRUG OFFENSE' in x.upper())]

grouped_non_violent_crime_df = non_violent_crime_df.groupby('UCR CRIME CATEGORY')


non_violent_crime_df
```

```python
#separating and grouping the violent crimes

#selecting the data that don't belong to non_violent_crime_df
violent_crime_df = crime_df[~crime_df.index.isin(non_violent_crime_df.index)]

grouped_violent_crime_df = violent_crime_df.groupby('UCR CRIME CATEGORY')


violent_crime_df
```
