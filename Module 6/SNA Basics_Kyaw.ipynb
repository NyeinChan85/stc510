```python
#importing pandas, matplotlib.pyplot and networkx libraries
import pandas as pd 
import matplotlib.pyplot as plt
import networkx as nx
```

```python
#creating a dataframe using the csv which contains Cocaine Trafficking Organization in New York (source: http://www.casos.cs.cmu.edu/tools/datasets/external/index.php#covert)
df = pd.read_csv('COCAINE_DEALING.csv', index_col=0)
df.head()
```

```python
#priting the number of rows and colomns that the df contains
print(df.shape)
```

```python
#summary of our dataframe
df.info()
```

```python
#creating a graph G using the dataframe: df
G = nx.Graph(df)
```

```python
print(G)
```

```python
#computing the betweenness centrality for each node
nx.betweenness_centrality(G)
```

```python
#calculating the eigenvector centrality for each node
nx.eigenvector_centrality(G)
```

```python
#calculating the average clustering coefficient
nx.average_clustering(G)
```

```python
#calculating centrality measures for nodes 
nx.degree_centrality(G)
```

```python
#computing positions for notes using the spring layout 
nx.spring_layout(G)
```

```python
#visualizing the newtwork
nx.draw_networkx(G)
```

```python
# the loop will iterate throgh each individual row of df, and for each row, it will iterates all the columns
for source, row in df.iterrows():
    for target, weight in row.iteritems():
        # checking whether there is a connection between the source and target nodes.
        if weight != 0:
            #adding egdes to the graph
            G.add_edge(source, target, weight=weight)
```

```python
#finding out degree centrality and printing it
degree_centrality = nx.degree_centrality(G)
```

```python
print(degree_centrality)
```

```python
#assigning the node with minimum centrality as the the weakest list 
weakest_link = min(degree_centrality, key=degree_centrality.get)

#analyzing the graph and visualizing the network
plt.figure(figsize=(13, 10))
pos = nx.spring_layout(G, seed=45)  

nx.draw_networkx(G, pos, node_size=200, width=0.5, font_size=10)

#determing the wealkest link and drawing it
nx.draw_networkx_nodes(G, pos, nodelist=[weakest_link], node_color='y', node_size=200)

#adding title to the plot and displaying the graph
plt.title('Covert Social Network - Cocaine Trafficking Organization in New York ')
plt.text(0.5, 0.95, '(the weakest link is highlighted in yellow)', horizontalalignment='center', fontsize=10, transform=plt.gca().transAxes)
plt.axis('off')
plt.show()

#creating a graph which is a copy of original graph
G_excluding_weakest_link = G.copy()

#removing the weakest link from the newly created graph
G_excluding_weakest_link.remove_node(weakest_link)

plt.figure(figsize=(13, 10))
pos_excluding_weakest_link = nx.spring_layout(G_excluding_weakest_link, seed=40)

nx.draw_networkx(G_excluding_weakest_link, pos_excluding_weakest_link, node_size=200, width=0.5, font_size=10)

#adding title to the plot and displaying the graph
plt.title('Covert Social Network - Cocaine Trafficking Organization in New York')
plt.text(0.5, 0.95, '(after the weakest link was removed)', horizontalalignment='center', fontsize=10, transform=plt.gca().transAxes)
plt.axis('off')
plt.show()
```

CONCLUSION
--------------------
Removing the weakest link in a covert social network has significant consequences. Firstly, it is noticeable that the nx.average_clustering coefficient has increased, indicating how the remaining nodes are well-connected after the weakest link was removed. Secondly, it divides the network into separate parts, cutting off nodes connected through the weakest link. This reduction in connectivity limits access to critical information for nodes reliant on the weakest link.

Without the weakest link, the network's structure changes, with some nodes gaining importance while others become less significant. Essentially, removing the weakest link alters communication, collaboration, and network structure, offering insights into resilience and vulnerabilities.
